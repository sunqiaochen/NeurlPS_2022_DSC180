```python
#Import modules and packages we need
import numpy as np
from scipy.io import wavfile
import librosa
import matplotlib.pyplot as plt
import pylab
import IPython.display as ipd
from glob import glob
import os
import pickle
import csv
import tensorflow
import librosa.display
import pandas as pd
```

```python
largeDataset = pd.read_csv('peru_large_dataset.csv')
largeDataset.head()
```

```python
largeDataset['AudioMothCode'].unique()
goodCode = ['AM-1', 'AM-2', 'AM-22', 'AM-24', 'AM-25', 'AM-26', 'AM-27', 'AM-29', 'AM-30', 'AM-6']
largeDataset = largeDataset[largeDataset['AudioMothCode'].isin(goodCode)]
largeDataset
```

```python
largeDataset = largeDataset[['AudioMothCode','StartDateTime','Global_Score','SourceFile']]
largeDataset = largeDataset[largeDataset['Global_Score'] > 0.5]
largeDataset.head()
```

```python
len(largeDataset)
```

```python
largeDataset.sort_values(['StartDateTime']).head()
```

```python
largeDataset['StartDateTime']= pd.to_datetime(largeDataset['StartDateTime'])
largeDataset['StartDateTime'] = largeDataset['StartDateTime'].dt.hour
sourceFile = largeDataset.groupby(['AudioMothCode', 'StartDateTime']).first()['SourceFile'].tolist()
sourceFile
```

```python
# generate dataset
result = pd.read_csv('peru_large_dataset.csv')
result = result[result['SourceFile'].isin(sourceFile)]
result.to_csv('method2.csv')
```
