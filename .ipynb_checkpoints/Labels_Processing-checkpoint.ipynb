```python
import pandas as pd
import matplotlib.pyplot as plot
```

```python
# Reading in all of the csv files
method1 = pd.read_csv("NeurlPS_Labels\stratified_random_sample_label(method1).csv")
method2 = pd.read_csv("NeurlPS_Labels\Microfaune_score_label(method2).csv")
method3 = pd.read_csv("NeurlPS_Labels\dawndusk(method3).csv")
method4 = pd.read_csv("NeurlPS_Labels\method4.csv")
```

```python
def new(method):
    new_header = method.iloc[0]
    method = method[1:]
    method.columns = new_header
    return method
```

```python
method1 = new(method1)
method1.drop(method1.tail(1).index,inplace=True)
method2 = new(method2)
method3 = new(method3)
method4 = new(method4)
```

```python
# Numbers for Method1
stratified_random_sample_clip_count = len(method1)
print("Total clips in Stratified Random Sample: ", str(stratified_random_sample_clip_count))

stratified_random_sample_bird_present_clip_count = len(method1[method1["Bird Vocalization"] == '1'])
print("Number of clips confirmed to have birds: ",str(stratified_random_sample_bird_present_clip_count))

stratified_random_sample_bird_absent_clip_count = len(method1[method1["Bird Vocalization"] == '0'])
print("Number of clips confirmed to not have birds: ",str(stratified_random_sample_bird_absent_clip_count))

stratified_random_sample_high_activity_clip_count = len(method1[method1["Heavy Activity"] == '1'])
print("Number of clips confirmed to have multiple birds: ",str(stratified_random_sample_high_activity_clip_count))

stratified_random_sample_low_activity_clip_count = len(method1[method1["Heavy Activity"] == '0'])
print("Number of clips confirmed to have little to no birds: ",str(stratified_random_sample_low_activity_clip_count))
```

```python
# Numbers for method2
skewed_sample_clip_count = len(method2)
print("Total clips in Skewed Random Sample: ", str(skewed_sample_clip_count))

skewed_random_sample_bird_present_clip_count = len(method2[method2["Bird Vocalization"] == "1"])
print("Number of clips confirmed to have birds: ",str(skewed_random_sample_bird_present_clip_count))

skewed_random_sample_bird_absent_clip_count = len(method2[method2["Bird Vocalization"] == "0"])
print("Number of clips confirmed to not have birds: ",str(skewed_random_sample_bird_absent_clip_count))

skewed_random_sample_high_activity_clip_count = len(method2[method2["Heavy Activity"] == "1"])
print("Number of clips confirmed to have multiple birds: ",str(skewed_random_sample_high_activity_clip_count))

skewed_random_sample_low_activity_clip_count = len(method2[method2["Heavy Activity"] == "0"])
print("Number of clips confirmed to have little to no birds: ",str(skewed_random_sample_low_activity_clip_count))
```

```python
# Numbers for method3
dawndusk_sample_clip_count = len(method3)
print("Total clips in Dawn-Dusk Random Sample: ", str(dawndusk_sample_clip_count))

dawndusk_random_sample_bird_present_clip_count = len(method3[method3["Bird Vocalization"] == "1"])
print("Number of clips confirmed to have birds: ",str(dawndusk_random_sample_bird_present_clip_count))

dawndusk_random_sample_bird_absent_clip_count = len(method3[method3["Bird Vocalization"] == "0"])
print("Number of clips confirmed to not have birds: ",str(dawndusk_random_sample_bird_absent_clip_count))

dawndusk_random_sample_high_activity_clip_count = len(method3[method3["Heavy Activity"] == "1"])
print("Number of clips confirmed to have multiple birds: ",str(dawndusk_random_sample_high_activity_clip_count))

dawndusk_random_sample_low_activity_clip_count = len(method3[method3["Heavy Activity"] == "0"])
print("Number of clips confirmed to have little to no birds: ",str(dawndusk_random_sample_low_activity_clip_count))
```

```python
# Numbers for Skewed Dawn-Dusk Random Sample
skewed_dawndusk_sample_clip_count = len(method4)
print("Total clips in Dawn-Dusk Random Sample: ", str(skewed_dawndusk_sample_clip_count))

skewed_dawndusk_random_sample_bird_present_clip_count = len(method4[method4["Bird Vocalization"] == "1"])
print("Number of clips confirmed to have birds: ",str(skewed_dawndusk_random_sample_bird_present_clip_count))

skewed_dawndusk_random_sample_bird_absent_clip_count = len(method4[method4["Bird Vocalization"] == "0"])
print("Number of clips confirmed to not have birds: ",str(skewed_dawndusk_random_sample_bird_absent_clip_count))

skewed_dawndusk_random_sample_high_activity_clip_count = len(method4[method4["Heavy Activity"] == "1"])
print("Number of clips confirmed to have multiple birds: ",str(skewed_dawndusk_random_sample_high_activity_clip_count))

skewed_dawndusk_random_sample_low_activity_clip_count = len(method4[method4["Heavy Activity"] == "0"])
print("Number of clips confirmed to have little to no birds: ",str(skewed_dawndusk_random_sample_low_activity_clip_count))
```
